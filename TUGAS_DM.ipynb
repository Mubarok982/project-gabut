```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('nba.xlsx')
df.head()
```

```python
# Menampilkan statistik deskriptif dari dataset
df.describe(include="all")
```

```python
# Membaca data dari sheet yang tersedia
df = pd.read_excel('nba.xlsx', sheet_name=0)

# Memilih kolom yang relevan
kolom_numerik = ["Age", "Weight"]

# Visualisasi Histogram
plt.figure(figsize=(12, 6))
df[kolom_numerik].hist(bins=20, alpha=0.7, figsize=(12, 6))
plt.suptitle("Histogram Persebaran Data")
plt.show()
```

```python
# Menghapus data Salary yang NaN
salary_data = df["Salary"].dropna()

# Visualisasi Histogram untuk Salary dengan lebih detail
plt.figure(figsize=(12, 6))
n, bins, patches = plt.hist(salary_data, bins=40, color='skyblue', edgecolor='black', alpha=0.7, density=True)

# Menyesuaikan rentang gaji agar lebih mudah dibaca
plt.xticks(np.linspace(salary_data.min(), salary_data.max(), num=10), rotation=45)

# Menambahkan garis distribusi KDE
sns.kdeplot(salary_data, color='red', linewidth=2, label="KDE")

# Menambahkan garis Mean dan Median
mean_salary = salary_data.mean()
median_salary = salary_data.median()
plt.axvline(mean_salary, color='green', linestyle='dashed', linewidth=2, label=f'Mean: {mean_salary:,.0f}')
plt.axvline(median_salary, color='blue', linestyle='dashed', linewidth=2, label=f'Median: {median_salary:,.0f}')

# Menambahkan label
plt.title("Histogram Salary - Distribusi Gaji Pemain NBA", fontsize=16)
plt.xlabel("Salary (USD)", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.legend()
plt.grid(axis='y', linestyle="--", alpha=0.7)

plt.show()
```

```python
# Memilih hanya kolom numerik
numeric_df = df.select_dtypes(include=['number'])

# Membuat Heatmap Korelasi
plt.figure(figsize=(10, 6))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap Korelasi antar Variabel")
plt.show()
```

## Filtering dengan kriteria gaj di angka 5-15 juta dan usia di antara 25-35 tahun

```python
# Filter data usia antara 25-35 dan gaji dalam rentang tertentu
df_filtered = df[(df['Age'] >= 25) & (df['Age'] <= 35) & (df['Salary'] >= 5000000) & (df['Salary'] <= 15000000)]

# Histogram distribusi gaji
plt.figure(figsize=(10, 6))
sns.histplot(df_filtered['Salary'], bins=20, kde=True, color='blue')
plt.title("Distribusi Gaji untuk Usia 25-35 dengan Gaji 5-15 Juta", fontsize=14)
plt.xlabel("Gaji", fontsize=12)
plt.ylabel("Frekuensi", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Menghitung rata-rata gaji berdasarkan posisi dan usia
gaji_mean = df_filtered.groupby(['Position', 'Age'])['Salary'].mean().reset_index()
print(gaji_mean)
```

## Filtering pemain dengan posisi PG, umur di atas 30 tahun, dan gaji di atas 12.000.000

```python
# Filtering pemain dengan posisi PG, umur di atas 30 tahun, dan gaji di atas 12.000.000
filtered_players = df[(df["Position"] == "PG") & (df["Age"] > 25) & (df["Salary"] > 12000000)]

# Menampilkan hasil filter
print(filtered_players)
```

```python
# Mengurutkan data dari gaji terendah ke tertinggi
filtered_players = filtered_players.sort_values(by="Salary")

# Visualisasi data dengan bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x=filtered_players["Name"], y=filtered_players["Salary"], palette="Blues_d")

# Menambahkan judul dan label
plt.title("Gaji Pemain PG dengan Usia >25 Tahun dan Gaji >12M", fontsize=12)
plt.xlabel("Nama Pemain", fontsize=10)
plt.ylabel("Gaji (USD)", fontsize=10)
plt.xticks(rotation=45)

# Menampilkan plot
plt.show()
```

## Sorting pemain dengan gaji tertinggi

```python
# Sorting pemain dengan gaji tertinggi
top_salary = df.nlargest(10, "Salary")

plt.figure(figsize=(12, 6))
sns.barplot(x="Salary", y="Name", data=top_salary, palette="coolwarm")
plt.xlabel("Gaji (USD)")
plt.ylabel("Nama Pemain")
plt.title(" 10 Pemain dengan Gaji Tertinggi di NBA")
plt.show()
```

## Distribusi Pemain Berdasarkan Posisi

```python
plt.figure(figsize=(8, 8))
df["Position"].value_counts().plot.pie(autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title(" Distribusi Pemain Berdasarkan Posisi")
plt.ylabel("")  # Hilangkan label default
plt.show()
```

## Melihat tren kenaikan gaji berdasarkan tahun pertama mereka bermain di NBA

```python
df_grouped = df.groupby("Start_Year")["Salary"].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x="Start_Year", y="Salary", data=df_grouped, marker="o", linewidth=2, color="b")
plt.xlabel("Tahun Masuk NBA")
plt.ylabel("Rata-rata Gaji (USD)")
plt.title(" Tren Kenaikan Gaji Berdasarkan Tahun Masuk NBA")
plt.grid(True)
plt.show()
```
